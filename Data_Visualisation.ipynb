**2. Cleaning and manipulating data**

Using the dataframe *df_movies*, peform the following:

*   Check whether there are any missing values in *df_movies*.
*   If yes, drop those values and print the size of df after dropping these.


```
# Check for missing values 
df_movies.isna().sum()
```

```
# Drop the missing values
movies = df_movies.dropna()
print(movies.shape)
```

```
movies['runtime'].describe()
```

```
df_movies.loc[df_movies['runtime'].isnull()]
```

Using the dataframe *df_movies*, peform the following:

*   Add another column named runtimeNew1 in *df_movies* by using **Min-Max Normalization** ,subtracting the minimum value from runtime and dividing it by (difference of Maximum value and Minimum Value = Range).

   **$(runtimeNew1)_i$** $ = ((runtime)_i- (min))/((max)-(min))$

```
# Add a column named runtimeNew1 
min = movies['runtime'].min()
max = movies['runtime'].max()
movies['runtimeNew1'] = (movies['runtime'] - min) / (max - min)
movies['runtimeNew1'].head(10)
```

Using the dataframe *df_movies*, peform the following:

*   Add another column named **runtimeNew2** in *df_movies* by using **Mean Std Normalization** ,subtracting the mean from runtime and dividing it by standard deviation..

**$(runtimeNew2)_i$**  $=((runtime)_i−(mean))/(standarddeviation)$



```
# Add a column named runtimeNew2 
mean = movies['runtime'].mean()
std_dev = movies['runtime'].std()
movies['runtimeNew2'] = (movies['runtime'] - mean) / std_dev
movies['runtimeNew2'].head(10)
```

```
df_movies['runtime'].mean()
```

```
# Replace the missing values 
df_movies['runtime'] = df_movies['runtime'].fillna(df_movies['runtime'].mean())
print(df_movies.isna().sum())
```
